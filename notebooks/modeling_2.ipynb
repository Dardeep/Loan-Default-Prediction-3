## Logistic Regression 

First we will focus on Logistic Regression, referring to the dataset as logistic_df, and using Xl and yl for features and target

```python
Xl = logistic_df.drop(columns=['loan_status'])
yl = logistic_df['loan_status']


Xl_train, Xl_test, yl_train, yl_test = train_test_split(Xl, yl, test_size=0.2, random_state=42, stratify=yl)


logistic_model = LogisticRegression(max_iter=1000, random_state=42)
logistic_model.fit(Xl_train, yl_train)


yl_pred = logistic_model.predict(Xl_test)
yl_prob = logistic_model.predict_proba(Xl_test)[:, 1]
print("Logistic Regression Performance:")
print(classification_report(yl_test, yl_pred))
print(f"AUC-ROC: {roc_auc_score(yl_test, yl_prob):.2f}")
```

#### Observation

- Overall Accuracy: 87%

     - Indicates the percentage of predictions (both default and non-default) that are correct.
    - Strong overall performance.

- Precision:

    - Class 0 (Non-default): 0.89
        - Out of all predicted non-default loans, 89% were correct.
    - Class 1 (Default): 0.79
        - Out of all predicted default loans, 79% were correct.

- Recall:

    - Class 0 (Non-default): 0.96
        - Very high recall means most actual non-default loans are correctly identified.
    - Class 1 (Default): 0.58
        - Lower recall indicates the model misses some default loans.

- F1-Score:

    - Balances precision and recall.
    - Class 0 (Non-default): 0.92
    - Class 1 (Default): 0.67

- AUC-ROC: 0.88

    - Indicates good discrimination between default and non-default loans.


- Imbalance Handling:

    - Class 0 (Non-default) has better metrics than Class 1 (Default), which is common when dealing with imbalanced data.
    - While precision for defaults (0.79) is good, recall (0.58) is relatively low, meaning some defaults are missed.

- Model Strengths:

    - The model performs well overall, especially for identifying non-default loans.

- Areas for Improvement:

    - The recall for defaults could be improved to reduce the risk of missing loans likely to default.

## Address Class Imbalance:

#### Model 1 we had used 3 different techniques and came to the conclusion that Class Weights was the best approach for Logistic Regression when dealing with our imbalanced dataset.

Since we created a function we will use that same code to see which technique is the best now that we have completed our feature engineering stage. 

```python
# Helper function to evaluate Logistic Regression
def evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]
    print(classification_report(y_test, y_pred))
    print(f"AUC-ROC: {roc_auc_score(y_test, y_pred_proba):.3f}")



# 1. Logistic Regression with Class Weights
print("Logistic Regression with Class Weights")
weighted_model = LogisticRegression(class_weight='balanced', random_state=42, solver='liblinear')
evaluate_model(weighted_model, Xl_train, Xl_test, yl_train, yl_test)

# 2. Logistic Regression with SMOTE
print("\nLogistic Regression with SMOTE")
smote = SMOTE(random_state=42)
Xl_train_smote, yl_train_smote = smote.fit_resample(Xl_train, yl_train)
oversampled_model = LogisticRegression(random_state=42, solver='liblinear')
evaluate_model(oversampled_model, Xl_train_smote, Xl_test, yl_train_smote, yl_test)

# 3. Logistic Regression with Undersampling
print("\nLogistic Regression with Undersampling")
undersampler = RandomUnderSampler(random_state=42)
Xl_train_under, yl_train_under = undersampler.fit_resample(Xl_train, yl_train)
undersampled_model = LogisticRegression(random_state=42, solver='liblinear')
evaluate_model(undersampled_model, Xl_train_under, Xl_test, yl_train_under, yl_test)
```

### Observation 
We can see that Class weights again is the best option, with slightly better AUC-ROC and recall for defaults compared to SMOTE and undersampling.

- AUC-ROC: 0.885
- Performance:
    - Good precision for Class 0 (Non-default): 0.93
    - Improved recall for Class 1 (Default): 0.79 (higher compared to the baseline model without weights).
    - Balanced class weighting helps the model focus more on the minority class.
- Accuracy: 0.82 (slightly lower than the baseline due to the trade-off with recall for Class 1).

# Hyperparameter tuning for Logistic Regression

We will now start with Hyperparameter tuning for Logistic Regression with class weights using GridSearchCV. This optimizes parameters such as C (regularization strength), solver, and class_weight. After finding the best parameters, the tuned model will be evaluated.

```python
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100], 
    'solver': ['liblinear', 'saga'],
    'class_weight': ['balanced', None]
    }

logistic_model = LogisticRegression(max_iter=1000, random_state=42)
grid_search = GridSearchCV(estimator=logistic_model, param_grid=param_grid, cv=3, scoring='roc_auc', verbose=2, n_jobs=-1)
grid_search.fit(Xl_train, yl_train)

best_model = grid_search.best_estimator_
print("Best Parameters:")
print(grid_search.best_params_)

def evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]
    print(classification_report(y_test, y_pred))
    print(f"AUC-ROC: {roc_auc_score(y_test, y_pred_proba):.3f}")

print("\nPerformance of Tuned Logistic Regression Model:")
evaluate_model(best_model, Xl_train, Xl_test, yl_train, yl_test)
```

### Result

##### Recall Improvement:
Recall for defaults (Class 1) improved from 0.58 (baseline Logistic Regression) to 0.79, showing the effectiveness of tuning and class weighting.
##### AUC-ROC Improvement:
The AUC-ROC increased from 0.88 (baseline) to 0.891, indicating better discrimination.

## Hyperparameter tuning for Gradient Boosting

We will now work on hyperparameter tuning for Gradient Boosting. This will optimize parameters such as n_estimators, learning_rate, max_depth, and subsample. Once we've found the best parameters, the tuned model will be evaluated on the test set. We will be using SMOTE as it had higher recall. 

```python
# SMOTE for handling imbalance
smote = SMOTE(random_state=42)
Xg_train_smote, yg_train_smote = smote.fit_resample(Xg_train, yg_train)

# Hyperparameter tuning for Gradient Boosting
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5, 7],
    'subsample': [0.8, 1.0]
}

gb_model = GradientBoostingClassifier(random_state=42)
grid_search = GridSearchCV(estimator=gb_model, param_grid=param_grid, cv=3, scoring='roc_auc', verbose=2, n_jobs=-1)
grid_search.fit(Xg_train_smote, yg_train_smote)

# Evaluation
best_model = grid_search.best_estimator_
print("Best Parameters:")
print(grid_search.best_params_)

# Evaluate the tuned model
def evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]
    print(classification_report(y_test, y_pred))
    print(f"AUC-ROC: {roc_auc_score(y_test, y_pred_proba):.3f}")

print("\nPerformance of Tuned Gradient Boosting Model:")
evaluate_model(best_model, Xg_train_smote, Xg_test, yg_train_smote, yg_test)
```

## Results

Gradient Boosting significantly outperforms Logistic Regression in terms of AUC-ROC, accuracy, and precision.

But  when it comes to recall for Class 1 (Defaults):
Logistic Regression (0.79) is slightly higher than Gradient Boosting (0.75), making it better at catching defaults.
Gradient Boosting compensates with much higher precision and overall robustness.

#### Other things to note:
Logistic Regression is significantly faster, in our case speed isn't a concern but with Logistic Regression having a higher recall that maybe a better option. 

#### Is there anything else we can do?
We can Adjust the Decision Threshold. 
- The default threshold for classifying a sample as default is 0.5 (i.e., probability â‰¥ 0.5 is classified as default).
- By lowering this threshold, more samples will be classified as defaults, increasing recall at the expense of precision.


```python
smote = SMOTE(random_state=42)
Xg_train_smote, yg_train_smote = smote.fit_resample(Xg_train, yg_train)

# Use the best parameters from GridSearchCV
best_model = GradientBoostingClassifier(
    n_estimators=300,
    learning_rate=0.1,
    max_depth=7,
    subsample=0.8,
    random_state=42
)

# Train the model
best_model.fit(Xg_train_smote, yg_train_smote)

# Evaluate the model at different thresholds
def evaluate_threshold(model, X_test, y_test, threshold):
    y_prob = model.predict_proba(X_test)[:, 1]
    y_pred = (y_prob >= threshold).astype(int)
    print(f"\nThreshold: {threshold}")
    print(classification_report(y_test, y_pred))
    print(f"AUC-ROC: {roc_auc_score(y_test, y_prob):.3f}")

# Plot Precision-Recall Curve and adjust threshold
y_prob = best_model.predict_proba(Xg_test)[:, 1]
precision, recall, thresholds = precision_recall_curve(yg_test, y_prob)

plt.figure(figsize=(8, 6))
plt.plot(thresholds, precision[:-1], label='Precision')
plt.plot(thresholds, recall[:-1], label='Recall')
plt.xlabel('Threshold')
plt.ylabel('Precision/Recall')
plt.title('Precision-Recall vs Threshold')
plt.legend()
plt.grid()
plt.show()

# Evaluate at specific thresholds
for threshold in [0.3, 0.4, 0.5]:
    evaluate_threshold(best_model, Xg_test, yg_test, threshold)
```

### Observation

##### Threshold = 0.3:

- Best Recall for Class 1 (Defaults): 81%.
- Recall improves significantly compared to the default threshold (0.5).
- Precision decreases slightly because more non-defaults are classified as defaults.

```python
final_threshold = 0.3
```

```python
_prob = best_model.predict_proba(Xg_test)[:, 1] 
y_pred_final = (y_prob >= final_threshold).astype(int)  


print(f"Performance with Finalized Threshold ({final_threshold}):")
print(classification_report(yg_test, y_pred_final))
print(f"AUC-ROC: {roc_auc_score(yg_test, y_prob):.3f}")
```
