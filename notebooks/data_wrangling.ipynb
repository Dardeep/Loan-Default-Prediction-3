### Cross Tables
We will now create a cross table for loan_intent and loan_status.

```python

print(pd.crosstab(loan_data['loan_intent'], loan_data['loan_status'], margins = True))
```

###### loan_intent vs loan_status
    crosstab shows the breakdown of loan_status (0 = non-default, 1 = default) across different loan_intent categories:

  <b>Debt Consolidation:</b>
1,490 defaults out of 5,212 loans (approx 28.6% default rate).

<b>Education:</b>
1,111 defaults out of 6,453 loans (approx 17.2% default rate).

<b>Home Improvement:</b>
941 defaults out of 3,605 loans (approx 26.1% default rate).

 <b>Medical:</b>
1,621 defaults out of 6,071 loans (approx 26.7% default rate).

<b>Personal:</b>
1,098 defaults out of 5,521 loans (approx 19.9% default rate).

<b>Venture:</b>
847 defaults out of 5,719 loans (approx 14.8% default rate).


##### Key Observations:
- The Debt Consolidation category has a high default count and rate, indicating higher risk.
- Education and Venture loans show lower default rates compared to other intents.
- Medical and Home Improvement loans also show notable default rates, similar to Debt Consolidation.

##### person_home_ownership, loan_grade, and loan_status
    This examines the interaction between person_home_ownership, loan_grade, and loan_status.

```python
print(pd.crosstab(loan_data['person_home_ownership'],[loan_data['loan_status'],loan_data['loan_grade']]))
```


<b>Mortgage:</b>
Majority of borrowers are in the Mortgage category, but they also exhibit higher counts of defaults across grades.
Defaults increase as the loan grade worsens (from A to G).

<b>Rent:</b>
Renters show the most defaults in lower loan grades (D to G), indicating higher risk in this group.

<b>Own:</b>
Defaults are fewer for borrowers who own their homes, especially in higher loan grades (A to C).

<b>Other:</b>
Smallest group with very few defaults, but defaults appear in lower grades.

<b>Loan Grade Pattern:</b>
Across all ownership types, defaults are significantly more frequent in lower loan grades (E, F, and G).

#### person_home_ownership and loan_percent_income by loan_status
    This shows the average loan_percent_income (loan amount as a percentage of income) grouped by person_home_ownership and loan_status.

```python
print(pd.crosstab(loan_data['person_home_ownership'], loan_data['loan_status'],
                  values=loan_data['loan_percent_income'], aggfunc='mean'))
```

<b>Observations:</b>

<b>Non-defaults</b>:
Borrowers across all homeownership types have relatively low loan_percent_income, ranging from ~0.14 to ~0.18.

<b>Defaults</b>:
Borrowers with defaults have significantly higher loan_percent_income, ranging from ~0.18 to ~0.30.
    OWN and OTHER homeownership types have the highest loan_percent_income for defaulters.

Borrowers in the RENT category show a notable increase in loan_percent_income when defaulting (0.14 to 0.26).


```python
loan_data.boxplot(column = ['loan_percent_income'], by = 'loan_status')
plt.title('Average Percent Income by Loan Status')
plt.suptitle('')
plt.show()
```

It looks like the average percentage of income for defaults is higher. This could indicate those borrowers have a debt-to-income ratio that's already too high.

### Dealing with Outliers
First we are going to deal with the age and the employment length since we've found a few outliers

```python
print(pd.crosstab(loan_data['loan_status'],loan_data['person_home_ownership'],
                  values=loan_data['person_emp_length'], aggfunc='max'))
```

```python
#Employment length is greater than 60 to be dropped
to_drop = loan_data[loan_data['person_emp_length'] > 60].index
```

```python
#Now we are going to drop the 
loan_data_new = loan_data.drop(to_drop)
```

```python
print(pd.crosstab(loan_data_new['loan_status'],loan_data_new['person_home_ownership'],
            values=loan_data_new['person_emp_length'], aggfunc=['min','max']))
```

```python
plt.scatter(loan_data_new['person_age'], loan_data_new['loan_amnt'], c='blue', alpha=0.5)
plt.xlabel("Person Age")
plt.ylabel("Loan Amount")
plt.show()
```

there are a few entries where the age is over 100. Let's go ahead and drop this

```python

loan_data = loan_data_new.drop(loan_data_new[loan_data_new['person_age'] > 100].index)

colors = ["blue","red"]
plt.scatter(loan_data['person_age'], loan_data['loan_int_rate'],
            c = loan_data['loan_status'],
            alpha=0.5)
plt.xlabel("Person Age")
plt.ylabel("Loan Interest Rate")
plt.show()
```

The above plot shows there are more defaults with high interest rates.

### Missing Data

```python
missing = loan_data.isnull().sum()
print(missing)
```

```python
loan_data[loan_data['person_emp_length'].isnull()].head()
```

```python
# We will be replacing the null values with the median value for all employment lengths
loan_data['person_emp_length'].fillna((loan_data['person_emp_length'].median()), inplace=True)
```

```python
n, bins, patches = plt.hist(loan_data['person_emp_length'], bins='auto', color='blue')
plt.xlabel("Person Employment Length")
plt.show()
```

For the missing loan int rate, we will be dropping the rows

```python
to_drop = loan_data[loan_data['loan_int_rate'].isnull()].index

loan_data_cleaned = loan_data.drop(to_drop)
```

```python
missing = loan_data_cleaned.isnull().sum()
print(missing)
```
