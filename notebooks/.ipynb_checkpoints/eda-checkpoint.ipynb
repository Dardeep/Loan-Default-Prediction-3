#### Categorical Features
As for the categorical features, we want to explore  the distribution of the categories and identify potential relationships with other variables like loan_status. Questions we want to explore are:

- Are certain loan intents (loan_intent) associated with higher default rates?
- Do loan_grade categories show a clear trend in defaults?
- Is there a significant difference in default rates based on person_home_ownership?
- Does cb_person_default_on_file (Y/N) correlate strongly with defaults?

```python
categorical_cols = ['loan_intent', 'loan_grade', 'person_home_ownership', 'cb_person_default_on_file']

for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col, palette="Set2")
    plt.title(f"Frequency Distribution of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
for col in categorical_cols:
    pd.crosstab(df[col], df['loan_status'], normalize='index').plot(
        kind='bar', stacked=True, figsize=(8, 5), colormap='viridis'
    )
    plt.title(f"Proportion of Defaults by {col}")
    plt.xlabel(col)
    plt.ylabel("Proportion")
    plt.legend(["Non-Default (0)", "Default (1)"], loc="upper right")
    plt.xticks(rotation=45)
    plt.show()
```

```python
for col in categorical_cols:
    print(f"Category Distribution for {col}:\n")
    print(df[col].value_counts(normalize=True) * 100)
    print("\n")
```

### Analysis of Categorical Variables

##### Loan Intent
Distribution:
The most common loan intents are Education (19.86%), Medical (18.50%), and Venture (17.48%), followed closely by Personal (17.01%).
Less frequent intents include Debt Consolidation (15.96%) and Home Improvement (11.17%).

Default Proportions:
Loans for Venture purposes show a significantly higher proportion of defaults compared to other intents, highlighting a riskier segment.
On the other hand, intents like Education and Debt Consolidation have relatively lower default rates, suggesting they are safer loan purposes.

##### Loan Grade
Distribution:
Loan grades are heavily concentrated in A (33.17%), B (31.88%), and C (19.78%), indicating a generally good credit quality distribution.
Lower grades such as D (11.25%), E (2.99%), F (0.73%), and G (0.20%) are much less frequent.

Default Proportions:
As expected, default rates increase with lower grades. Borrowers with grades F and G show exceptionally high default proportions, while grades A and B are associated with lower risks.

##### Home Ownership
Distribution:
The majority of borrowers are either Renters (50.54%) or have a Mortgage (41.14%). Only a small percentage Own (8.00%) their homes, with negligible representation for Other (0.32%).

Default Proportions:
Renters exhibit a notably higher default rate compared to borrowers with a mortgage or those who own their homes. This could reflect income instability or lower financial security among renters.

##### Default on File
Distribution:
The majority of borrowers (82.31%) have no history of default (N), while 17.69% have a prior default (Y).

Default Proportions:
Borrowers with a history of default (Y) show significantly higher default rates, reinforcing the importance of this feature in predicting loan outcomes.

#### Key Takeaways
Loan Intent: Certain purposes, like Venture, pose higher risks, whereas intents like Education and Debt Consolidation are safer.
Loan Grade: A borrowers loan grade strongly correlates with default rates, with higher grades (A, B) showing lower risks.

Home Ownership: Borrowers who rent are more likely to default compared to homeowners or those with a mortgage.
Default History: A history of default is a strong indicator of future defaults.

### One-Hot Encoding Categorical Variables

```python
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

df_encoded.head()
```

#### Correlation Matrix

```python
corr_matrix = df_encoded.corr()

plt.figure(figsize=(20, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Feature Correlations', size=15)
plt.show()
```

```python
target_corr = corr_matrix['loan_status'].sort_values(ascending=False)
print("\nCorrelation with Target Variable (loan_status):")
print(target_corr)
```

## Feature Correlation

### Variables Most Correlated with Loan Default (loan_status):

<b>loan_percent_income (0.38):</b> Higher loan payments as a percentage of income are strongly associated with defaults.

<b>loan_int_rate (0.34):</b> Higher interest rates are correlated with a higher likelihood of default.

<b>loan_grade_D (0.32), loan_grade_E (0.18), loan_grade_F (0.10), loan_grade_G (0.08):</b> Lower loan grades (indicative of riskier loans) have a positive correlation with default.

<b>person_home_ownership_RENT (0.24):</b> Renters are more likely to default compared to homeowners or those with mortgages.

<b>cb_person_default_on_file_Y (0.18):</b> Having a history of default is positively correlated with defaulting again.

<b>loan_amnt (0.11):</b> Larger loan amounts are slightly correlated with default.

### Variables Negatively Correlated with Loan Default:
<b>person_income (-0.17):</b> Higher income is associated with a lower likelihood of default.

<b>person_home_ownership_OWN (-0.10):</b> Homeownership is negatively correlated with default.

<b>loan_intent_EDUCATION (-0.06):</b> Loans taken for education are less likely to default compared to other purposes.

### Observations on Relationships Between Variables:
The loan grades (A to G) are strongly correlated with both interest rate and default, which makes sense as lower grades correspond to riskier loans with higher rates and higher default probabilities.
Income (person_income) and loan percent income are inversely related, as expected (higher income reduces the impact of loan payments)
