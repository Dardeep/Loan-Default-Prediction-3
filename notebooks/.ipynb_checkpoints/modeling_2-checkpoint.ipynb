# Loan Default Prediction - Modeling 2

### This notebook is after completing our Feature Engineering

#### Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from imblearn.over_sampling import SMOTE

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.metrics import roc_auc_score, classification_report, confusion_matrix

```

```python
df = pd.read_csv("../data/loan_data_fe_complete.csv", index_col=0)

df.head()
```

### Training

```python
X = df.drop('Default', axis=1)  
y = df['Default']  


X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=77, stratify=y)
```

<b> note to self</b>

stratify=y parameter in train_test_split() ensures that the class distribution of the target variable (y) is preserved in both the training and test sets. This is particularly useful when you're dealing with imbalanced datasets, as it ensures that both the training and test sets have a similar ratio of the majority and minority classes.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
X_train_s.shape
```

### Modeling - Keras

We are going to add dropout to prevent overfitting after 2 layers and output layer for binary classification. 

```python
model = Sequential([
    Dense(256, activation="relu", input_shape=[23]),
    Dropout(0.3),  
    Dense(256, activation="relu"),
    Dropout(0.3), 
    Dense(1, activation="sigmoid") 
])
```

```python
model.summary()
```

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

history = model.fit(X_train_s, y_train, validation_data=(X_test_s, y_test), batch_size=256, epochs=100, verbose=1)

y_pred_prob = model.predict(X_test_s)
y_pred_class = (y_pred_prob > 0.5).astype("int32")


```

```python
# Evaluate the model
auc = roc_auc_score(y_test, y_pred_prob)
print(f"AUC-ROC Score: {auc}")

print("Classification Report:\n", classification_report(y_test, y_pred_class))

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_class))
```

```python
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()

# Plot loss
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

## Keras Model Performance

As we evalueate the model the result were:

<b>AUC-ROC Score: 0.852</b> - A score of 0.85 means that the model has an 85% probability of correctly distinguishing between defaults vs. non-defaults.


<b>Precision:</b> 86% of the instances predicted as non-defaults are correct and 71% of the instances predicted as defaults correctly.
<b>Recall:</b> Model correctly identifies 63% of the actual non-defaults and correctly identifies 90% of the actual defaults.
<b>F1-score: </b> 0.73, indicating the balance between precision and recall for non-defaults and 0.79 for defaults.

### The model does an excellent job at identifying defaults, as seen by the high recall of 0.90. This is crucial when predicting defaults, as it's often more important to catch as many defaults as possible, even at the cost of some false positives.

The model struggles more with identifying non-defaults, with a recall of 0.63. This suggests that the model is more conservative, preferring to classify more instances as defaults to avoid missing actual defaults, which explains the relatively high number of false positives 

### What we can do to improve our Keras model:

1) Adjust Threshold -  We can try adjusting the threshold for the final classification to see if you can improve precision and recall for non-defaults.
2) Fine-Tune Model - We can continue to tune hyperparameters like the number of neurons, dropout rates, learning rate, or batch size to further improve model performance.

# Pytorch

Now let's build a similar model in PyTorch and compare its performance to the Keras model. PyTorch provides more control and flexibility in how you define and train your model, but the general architecture can be very similar to what you used in Keras.

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

#### Model

We will do it similar to Keras, have the layers with 23 features, 256 neurons and Dropout set to 0.3

```python
class LoanDefaultNN(nn.Module):
    def __init__(self):
        super(LoanDefaultNN, self).__init__()
        self.fc1 = nn.Linear(23, 256)  
        self.dropout1 = nn.Dropout(0.3)
        self.fc2 = nn.Linear(256, 256)  
        self.dropout2 = nn.Dropout(0.3)
        self.fc3 = nn.Linear(256, 1) 
    
    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.dropout1(x)
        x = torch.relu(self.fc2(x))
        x = self.dropout2(x)
        x = torch.sigmoid(self.fc3(x)) 
        return x


model = LoanDefaultNN()
```

```python
criterion = nn.BCELoss() 
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
X_train_tensor = torch.tensor(X_train_s, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.to_numpy(), dtype=torch.float32).view(-1, 1) 

X_test_tensor = torch.tensor(X_test_s, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.to_numpy(), dtype=torch.float32).view(-1, 1) 

epochs = 100
batch_size = 256

for epoch in range(epochs):
    model.train()
    
    permutation = torch.randperm(X_train_tensor.size()[0]) 
    # MINIBATCH TRAINING, REFER TO AUG 29TH NOTE NOTION 12.8.4
    # DO NOT REMOVE, IF THIS FAILS JUST MAKE IT A COMMENT. 
    
    for i in range(0, X_train_tensor.size()[0], batch_size):
        optimizer.zero_grad() # THIS IS USED TO CLEAR THE GRADIENTS. REFER TO MEMO NOTION 12.8.1
        indices = permutation[i:i + batch_size]
        batch_X, batch_y = X_train_tensor[indices], y_train_tensor[indices]
        

        outputs = model(batch_X)
        loss = criterion(outputs, batch_y)

        loss.backward()
        optimizer.step()

    if (epoch+1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}') #NOTE: THIS IS TO PRINT LOSS AT EVERY 10 EPOCH. 
```

#### Evaluate the model

note to self: Need to use model.eval() and convert 

```python

model.eval()
with torch.no_grad():
    y_pred_prob = model(X_test_tensor).numpy()


y_pred_class = (y_pred_prob > 0.5).astype(int)


auc = roc_auc_score(y_test, y_pred_prob)
print(f"AUC-ROC Score: {auc}")

print("Classification Report:\n", classification_report(y_test, y_pred_class))

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_class))
```
