## Review Numerical Columns

```python
#Understand Numeric Columns for our Data
print(loan_data.describe())
```

```python
#Default Variable Visualized

ax = loan_data['Default'].value_counts().plot(kind='bar',
                                    figsize=(14,8),
                                    title="Loans Marked as Defaults")
ax.set_xlabel("In good standing (0) vs Defaulted (1)")
ax.set_ylabel("Count")
```

From the describe() output, we're getting a good overview of our numerical data. 

<b>Age</b>: The mean age is 43, with a range from 18 to 69. This looks like a reasonable spread of ages, with no extreme outliers.

<b>Income</b>: The income has a large range, from $15,000 to $149,999, with an average of about $82,499. This wide distribution may impact how loans are granted or defaulted on, and it may be worth exploring the distribution further to identify any skewness or outliers.

<b>LoanAmount</b>: The loan amounts range from $5,000 to $249,999, with a median of $127,556. We might want to log-transform this column if we find that its distribution is highly skewed.

<b>CreditScore</b>: The mean credit score is 574, which is considered subprime, and the range extends from 300 to 849. We can explore the relationship between this feature and default rates, as it’s likely highly predictive of default risk.

<b>MonthsEmployed</b>: There's a large standard deviation here (34.64), indicating that employment history is quite varied. The max value of 119 months (about 10 years) suggests that there might be different levels of job security or career stability among the borrowers.

<b>NumCreditLines</b>: This feature has a relatively low variance, with most people having between 2 and 4 credit lines. Since the range is narrow, we may want to check if this feature has significant predictive power.

<b>InterestRate</b>: Ranges from 2% to 25%, with an average of 13.5%. The large standard deviation (6.6%) suggests a broad spread of interest rates, which could be influenced by creditworthiness and loan term.

<b>LoanTerm</b>: Ranges from 12 to 60 months, with a median of 36 months. This indicates that loan terms vary significantly, and it could be another important factor to assess.

<b>DTIRatio (Debt-to-Income Ratio)</b>: The ratio is fairly low on average (mean of 0.5), with a tight range from 0 to 0.9. This is a key factor to monitor, as a higher DTI often correlates with higher default risk.

<b>Default</b>: This is the target variable (binary classification), and about 11.6% of loans are marked as defaults. This class imbalance (around 1:9 ratio of defaults to non-defaults) may need to be addressed using techniques like SMOTE or adjusting the class weights during model training.

## Categorical Data


To start we are going to drop LoanID

```python
loan_data = loan_data.drop('LoanID', axis=1)
```

```python
loan_data.head(5)
```

###### Now we initialize Label Encoder for the categorical data

```python
le = LabelEncoder()

categorical_columns = ['Education', 'EmploymentType', 'MaritalStatus', 
                       'HasMortgage', 'HasDependents', 'LoanPurpose', 'HasCoSigner']

label_mappings = {}

for col in categorical_columns:
    loan_data[col] = le.fit_transform(loan_data[col])
    label_mappings[col] = dict(zip(le.classes_, le.transform(le.classes_)))


for col, mapping in label_mappings.items():
    print(f"Label encoding for {col}: {mapping}")
```

```python
loan_data.head(5)
```

<b>Label encoding for Education: </b>{"Bachelor's": 0, 'High School': 1, "Master's": 2, 'PhD': 3}

<b>Label encoding for EmploymentType:</b> {'Full-time': 0, 'Part-time': 1, 'Self-employed': 2, 'Unemployed': 3}

<b>Label encoding for MaritalStatus:</b> {'Divorced': 0, 'Married': 1, 'Single': 2}

<b>Label encoding for HasMortgage:</b> {'No': 0, 'Yes': 1}

<b>Label encoding for HasDependents:</b> {'No': 0, 'Yes': 1}

<b>Label encoding for LoanPurpose:</b> {'Auto': 0, 'Business': 1, 'Education': 2, 'Home': 3, 'Other': 4}

<b>Label encoding for HasCoSigner:</b> {'No': 0, 'Yes': 1}
